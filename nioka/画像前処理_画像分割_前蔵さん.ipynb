# 画像の分割(1024×512→128×128)

```python
path_C5 = glob.glob("./image/C5_raw/*.png")
path_F4 = glob.glob("./image/F4_raw/*.png")

for index, img_path in enumerate(path_C5):
    img = cv2.imread(img_path, 0)
    for i in range(8):
        for j in range(4):
            split_img = img[128*j:128*(j+1), 128*i:128*(i+1)]
            cv2.imwrite("./image/C5_split/" + str(index) + "-" + str(i) + "-"+ str(j) + ".png", split_img)

for index, img_path in enumerate(path_F4):
    img = cv2.imread(img_path, 0)
    for i in range(16):
        for j in range(8):
            split_img = img[128*j:128*(j+1), 128*i:128*(i+1)]
            cv2.imwrite("./image/F4_split/" + str(index) + "-" + str(i) + "-"+ str(j) + ".png", split_img)
```

# 画像の選択(ピクセルの閾値処理)

```python
path_C5 = glob.glob("./image/split_C5/*.png")
path_F4 = glob.glob("./image/split_F4/*.png")

for index, img_path in enumerate(path_C5):
    count = 0
    img = cv2.imread(img_path, 0)
    pix = img.ravel()
    
    for i in range(len(pix)):
        
        if(pix[i] <= 30):
            count += 1
            
    if(count >= 2000):
        shutil.move(img_path, "./image/C5_trash_image/")

for index, img_path in enumerate(path_F4):
    count = 0
    img = cv2.imread(img_path, 0)
    pix = img.ravel()
    
    for i in range(len(pix)):
        
        if(pix[i] <= 30):
            count += 1
            
    if(count >= 2000):
        shutil.move(img_path, "./image/F4_trash_image/")
```
